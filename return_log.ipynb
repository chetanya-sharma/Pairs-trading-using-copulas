```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('percentage_change.csv')
```

```python
for stock in df.columns:
    df[stock]=np.log(1+df[stock])

df
```

```python
for stock in df.columns:
    df[stock]=np.exp(df[stock].cumsum())-1

df
```

```python
df.to_csv('cumulative_log_returns.csv')
```
