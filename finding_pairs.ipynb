```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cumulative_log_returns.csv')
cutoff=int(len(df)*0.5)
df1=df.iloc[:-cutoff]
df1.drop(columns={f'{df1.columns[0]}'},inplace=True)
df1
```

```python
# First let us find out the correlation map of all the stocks present in the percentage_change.csv
# We are using spearman correlation to find the correlation between stocks
corr_matrix=df1.corr(method='spearman')
print(corr_matrix)
```

```python
corr_matrix.to_csv('corr_matrix.csv')
```

```python
df=pd.read_csv('corr_matrix.csv')
df.rename(columns={df.columns[0]:'stocks'},inplace=True)
df
```

```python
df.set_index('stocks',inplace=True)
df
```

```python
df.columns = df.columns.str.replace(' ', '')
df.index = df.index.str.replace(' ', '')
```

```python
pairs_list=[]
for row_index, row in df.iterrows():
    for col_name in df.columns:
        if row[col_name] > 0.85 and row_index!=col_name:
            if (row_index,col_name) not in pairs_list:
                pairs_list.append((col_name, row_index))

pairs_list
```

```python
df1=pd.DataFrame(pairs_list,columns=['Pair1','Pair2'])

df1.to_csv("pairs.csv")
```
