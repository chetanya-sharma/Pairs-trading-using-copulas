```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
```

```python
# first we generate the Q-Q or Quantile-Quantile plot for the 2 stocks chosen i.e. AARTIIND and ALKYLAMINE
df=pd.read_csv('Adjusted_Close.csv')
cutoff=int(len(df)*0.5)
df=df[:-cutoff]
df
```

```python
one='CARBORUNIV'
two='3MINDIA'
sorted_1=np.sort((df[f'{one}']-df[f'{one}'].min())/(df[f'{one}'].max()-df[f'{one}'].min()))
sorted_2=np.sort((df[f'{two}']-df[f'{two}'].min())/(df[f'{two}'].max()-df[f'{two}'].min()))
```

```python
plt.figure()
plt.scatter(sorted_1,sorted_2)
plt.plot([sorted_1[0],sorted_1[-1]],[sorted_2[0],sorted_2[-1]],color='b',ls='--')
plt.title('Q-Q plot: Time Series 1 vs Time Series 2')
plt.xlabel('Quantiles of Time Series 1')
plt.ylabel('Quantiles of Time Series 2')
plt.show()
```
